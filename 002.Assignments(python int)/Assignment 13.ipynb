```python
#1. Find all the words which contains 'uu' in the file (sowpods.txt)
file = open('sowpods.txt','r')
words = file.readlines()
for w in words:
    if 'UU' in w:
        print(w.strip())
file.close()
```

```python
#2. Print all alphabets that never appear double in sequence 
#(back to back) in the words in the files.
import string
import time
file = open('sowpods.txt','r')
words = file.readlines()
start = time.time()
for alpha in string.ascii_uppercase:
    flag = False
    for w in words:
        if (alpha*2) in w:
            flag = True
            break
    if flag == False:
        print(alpha)
print(time.time()-start)
file.close()
```

```python
#3. Print words in the file contains all the vowels.
vowels = 'AEIOU'
file = open('sowpods.txt','r')
words = file.readlines()
words = [i.strip() for i in words]
flag = True
for w in words:
    flag = 0
    for i in vowels:
        if i in w:
            flag += 1
    if(flag == 5):
        print(w)
file.close()
```

```python
#4. Print all palindrome words in the file.
def palindrome(word):
    if word == word[::-1]:
        return True
    return False
#print(palindrome('dfd'))
file = open('sowpods.txt','r')
words = file.readlines()
words=[i.strip() for i in words]
for w in words:
    if palindrome(w):
        print(w)
file.close()
```

```python
#5. Print the longest palindrome in sowpods.txt
def palindrome(word):
    if word ==word[::-1]:
        return True
    return False
longest=' '
file = open('sowpods.txt','r')
words = file.readlines()
words=[i.strip() for i in words]
for w in words:
    if palindrome(w):
        #print(w)
        if(len(w) > len(longest)):
            longest = w
print(longest)
file.close()
```

```python
#6. Print the words that are in sonnet_words.txt but not in sowpods.txt
import time
file = open('sowpods.txt','r')
words = file.readlines()
words=[i.strip() for i in words]
set_words={i.strip() for i in words}

s_file = open('sonnet_words.txt','r')
s_words = s_file.readlines()
s_words=[i.strip() for i in s_words]

start=time.time()
for sw in s_words:
    if sw not in set_words:
        print(sw)
file.close()
print(time.time()-start)
```

```python
#2)Write a Python program to count the frequency of words in a file.
from collections import Counter
def word_count(fname):
        with open(fname) as f:
                return Counter(f.read().split())

print("Number of words in the file : \n",word_count("sonnet_words.txt"))
```
