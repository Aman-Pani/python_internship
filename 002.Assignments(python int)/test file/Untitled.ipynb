```python
def login_module():
    print()
    print('#**Welcome to Employee System**#')
    Lid = input("Enter the login ID :: ")
    pas = input("Enter the password :: ")
    fa = open("login.txt","r")
    s = fa.readlines()
    c = 0
    for i in s:
        for j in i.split("\n"):
            if j == Lid + " " + pas and Lid == "admin":
                print("___Welcome to Admin module___")
                c += 1
                import admin
                return 1,Lid
            elif j == Lid + " " + pas:
                if search_1(Lid) == True:
                    print("___Welcome to HR module___")
                    c = c + 1
                    return 2,Lid,pas
                elif search_1(Lid) == False:
                    print("___Welcome to Employee module___")
                    c = c + 1
                    return 3,Lid,pas
    if c == 0:
        return 4
    fa.close()
          
def Add_employee(): # Adding new employes details function
    print()
    f1 = open("EMP.txt","a")
    fa = open("login.txt","a")
    a = ["Employee id :: ","Employee Name :: ","Employee DOJ :: ","Employee Designation :: ","Employee Salary :: "]
    print("Enter details of the employee :: ")
    b = []
    for j in a:
        print(j,end = "")
        g = input()
        b.append(g)
        f1.write(g)
        f1.write(" ")
    f1.write("\n")
    for i in range(0,2):
        fa.write(b[i])
        fa.write(" ")
    fa.write("\n")
    fa.close()
    f1.close()
    print("New Employee Added successfully !!!! ")
def Add_HR(): # Adding new HR details function
    print()
    f2 = open("HR.txt","a")
    a = ["Employee id :: ","HR department :: ","HR role :: "]
    print("Enter details :: ")
    c = 0
    b = []
    for j in a:
        print(j,end = "")
        c = input()
        b.append(c)
        f2.write(c)
        f2.write(" ")
    f2.write("\n")
    f2.close()
    if search(b[0]) == True:
        print("HR added successfully.")
    elif search(b[0]) == False:
        print("No such Employee exhist.")
        remove(b[0],"HR.txt"," ")
    print()
def see_employee_file():  # view employee file function
    print()
    print("Details of employee :: ")
    f1 = open("EMP.txt","r")
    r = f1.read()
    print(r)
    f1.close()
def see_hr_file():   # view HR file function
    print()
    print("Details of HR's :: ")
    f2 = open("HR.txt","r")
    r = f2.read()
    print(r)
    f2.close()
def search_employee(s_e):   # Searching employee function
    print()
    f1 = open("EMP.txt","r")
    s = f1.readlines()
    print("Status :: ",end = "")
    counter = 0
    for i in s:
        for j in i.split():
            if s_e == j:
                print("Employee found")
                print("Details of the searched employee is ::",i)
                counter += 1
                break
    if counter == 0:
        print("Employee not found . ")
    f1.close()
def search(emp_id):
    f1 = open("EMP.txt","r")
    s = f1.readlines()
    counter = 0
    for i in s:
        for j in i.split():
            if emp_id == j:
                counter += 1
                return True
                break
    if counter == 0:
        return False
    f1.close()
def search_1(emp_id):
    f2 = open("HR.txt","r")
    s = f2.readlines()
    counter = 0
    for i in s:
        for j in i.split():
            if emp_id == j:
                counter += 1
                return True
                break
    if counter == 0:
        return False
    f2.close()

def search_HR(s_HR): # searching in HR file Function 
    print()
    f2 = open("HR.txt","r")
    s = f2.readlines()
    counter = 0
    print("Status :: ",end = "")
    for i in s:
        for j in i.split():
            if s_HR == j:
                print("Employee found")
                print("Details of the searched employee is :: ",i)
                counter += 1
                break
    if counter == 0:
        print("Employee not found . ")
    f2.close()
def remove_employee(emp_id):
    if search(emp_id) == True: 
        remove(emp_id, 'EMP.txt', ' ')
        remove(emp_id, 'HR.txt', ' ')
        remove(emp_id, 'login.txt', ' ')
    elif search(emp_id) == False:
        print("Employee not found .")
    print()
def remove(emp_id, files, s):
    f = open(files, 'r')
    a = f.readlines()
    for i in range(len(a)):
        j = a[i].split(s)
        if j[0] == emp_id:
            a[i] = ''
            break
    f.close()
    f = open(files, 'w')
    f.writelines(a)
    f.close()


def admin_menu():
    c = True
    while c == True:
        o = input(
            "1.Add Employee\n2.Add HR\n3.See Employee file" +
            "\n4.See HR file\n5.search an Employee\n6.Search an HR" + 
            "\n7.Remove an Employee\n8.Remove HR\n9.Exit\nYour option >> ")
        if o == "1":
            Add_employee()
            print()
        elif o == "2":
            Add_HR()
            print()
        elif o == "3":
            see_employee_file()
            print()
        elif o == "4":
            see_hr_file()
            print()
        elif o == "5":
            s_e = input("Enter the Employee ID :: ")
            search_employee(s_e)
            print()
        elif o == "6":
            s_HR = input("Enter the ID :: ")
            search_HR(s_HR)
            print()
        elif o == "7":
            print()
            emp_id = input(("Enter the employee id :: "))
            remove_employee(emp_id)
        elif o == "8":
            print()
            emp_id = input(("Enter the employee id :: "))
            if search(emp_id) == True:
                remove(emp_id,"HR.txt"," ")
            elif search(emp_id) == False:
                print("No such Employee found.")
                print()
        else:
            c = False  
def designation_search():
    print()
    des = input("Enter the designation or type (ALL) to see all employees :: ")
    if des == "ALL" :
        see_employee_file()
    else:
        print("Details of the employees are :: ")
        f1 = open("EMP.txt","r")
        s = f1.readlines()
        counter = 0
        for i in s:
            for j in i.split():
                if des == j:
                    print(i)
                    counter += 1
                    break
        if counter == 0:
            print("Employee not found . ")
        f1.close()

def HR_menu(emp_id):
    c = True
    while(c == True):
        o = input("1.See your details\n2.view employees\n3.Exit\nyour choise >> ")
        if o == "1":
            print()
            search_employee(emp_id)
        elif o == "2":
            designation_search()
            print()
        elif o == "3":
            c = False

def Employee_menu(emp_id):
    c = True
    while(c == True):
        o = input("1.See your details\n2.View all HR details.\n3.Exit\noption >> ")
        if o == "1":
            print()
            search_employee(emp_id)
        elif o == "2":
            print()
            see_hr_file()
        elif o == "3":
            c = False
__name__ = "__main__"   # main functionS
while True:
    m,n,g = login_module()
    if  m == 1:
        print()
        admin_menu()
    elif m == 2:
        print()
        print(f"Welcome {g}")
        HR_menu(n)
    elif m == 3:
        print()
        print(f"Welcome {g} ")
        Employee_menu(n)
    elif m == 4:
        print("Invalid.")
```
